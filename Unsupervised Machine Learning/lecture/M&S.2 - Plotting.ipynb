## Matplotlib OO interface 
We use an instance of axes.Axes in order to render visualizations on an instance of figure.Figure.

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x,y)
```

## Working with DataFrame

```python
cars = pd.read_csv('../samples/mtcars.csv')
fig = plt.figure()
ax = fig.add_axes([.1, .1, 2, 2])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.iloc[:,0:1].values, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

## Seaborn Joinplot & Boxplot

```python
sb.jointplot( x=cars['cyl'], y=cars['mpg'] )
```

```python
sb.boxplot( x=cars['cyl'], y=cars['mpg'] )
```
