## Matplotlib OO interface 
We use an instance of axes.Axes in order to render visualizations on an instance of figure.Figure.

### The Figure Object

```python
fig = plt.figure()
fig.suptitle("Empty figure")
plt.show(block=False)
```

### The Axes Object
One or more axes objects are required to get us started with plotting.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x,y)
plt.show()
```

#### add_subplot(num_rows, num_cols, subplot_location)

creates a grid of of subplots of size (num_rows x num_cols) and returns an axes object for the subplot at subplot_location. 

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y)
ax2 = fig.add_subplot(1,2,2)
ax2.plot(x,y)
plt.show(block=False)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('X') # X Label
ax.set_ylabel('Y') # Y Label
ax.set_title('Simple XY plot') # Tile
ax.scatter(x,y)
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

### Joinplot
Displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins.

```python
sns.jointplot( x=cars['cyl'], y=cars['mpg'] )
```

###  Boxplot

```python
sns.boxplot( x=cars['cyl'], y=cars['mpg'] ) # Min, Max, Mean, Q1, Q3
```

### Pairplots

```python
sns.pairplot(cars)
```

### Correlation Heatmap

```python
# Correlation Matrix
cars.corr() # 1 positive corr, 0 no corr, -1 negative corr
```

```python
# Heatmap
sns.heatmap(cars.corr())
```

```python
sns.heatmap(abs(cars.corr()))
```
