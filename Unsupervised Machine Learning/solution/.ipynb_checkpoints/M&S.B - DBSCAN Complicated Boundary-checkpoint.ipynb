## DBSCAN for Complicated Boundary

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

import sklearn
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

from collections import Counter
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 7, 4
```

```python
moon_data, moon_target = make_moons(200, noise=.05)
plt.scatter(moon_data[:,0],moon_data[:,1])
```

### Apply DBSCAN

```python
model = DBSCAN(eps = 0.4, min_samples = 19).fit(moon_data)
print model
```

```python
print Counter(model.labels_)
```

```python
plt.scatter(moon_data[:,0],moon_data[:,1], c=model.labels_)
```
