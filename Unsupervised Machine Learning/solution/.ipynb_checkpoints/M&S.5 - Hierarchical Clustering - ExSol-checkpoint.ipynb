```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams

import sklearn

import scipy #for dendograms
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 7, 4
```

## Sample Data


```python
from sklearn.datasets.samples_generator import make_blobs

sample_data, sample_target = make_blobs(n_samples=50, centers=5, cluster_std=0.60, random_state=0)

color_theme = np.array(['red','grey','blue','black','yellow'])

plt.scatter(sample_data[:, 0], sample_data[:, 1],  c=color_theme[sample_target], s=50);
```

## Generate Linkage Matrix for Different Methods and Metrics

```python
from scipy.cluster.hierarchy import dendrogram, linkage

zw = linkage(sample_data, 'ward')
dendrogram(zw,  p=12, leaf_rotation=45., leaf_font_size=10.)
plt.title('Truncated Hierarchical Clustering Dendrograms')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

plt.show()
```

```python
zc = linkage(sample_data, 'complete')
dendrogram(zc,  p=12, leaf_rotation=45., leaf_font_size=10.)
plt.title('Truncated Hierarchical Clustering Dendrograms')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

plt.show()
```

```python
za = linkage(sample_data, 'average') 
dendrogram(za,  p=12, leaf_rotation=45., leaf_font_size=10.)
plt.title('Truncated Hierarchical Clustering Dendrograms')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

plt.show()
```

```python
zs = linkage(sample_data, 'single') 
dendrogram(zs,  p=12, leaf_rotation=45., leaf_font_size=10.)
plt.title('Truncated Hierarchical Clustering Dendrograms')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

plt.show()
```
