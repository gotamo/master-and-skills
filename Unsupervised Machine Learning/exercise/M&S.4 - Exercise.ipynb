```python
import numpy as np
import pandas as pd
import seaborn as sb;

import matplotlib.pyplot as plt
from matplotlib import rcParams

import sklearn
from sklearn import datasets

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.cluster import AgglomerativeClustering

import scipy #for dendograms
from scipy.cluster.hierarchy import dendrogram, linkage

from sklearn.datasets.samples_generator import make_blobs
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 7, 4
```

## Sample Data


```python
sample_data, sample_target = make_blobs(n_samples=50, centers=5, cluster_std=0.60, random_state=0)

color_theme = np.array(['red','grey','blue','black','yellow'])

plt.scatter(sample_data[:, 0], sample_data[:, 1],  c=color_theme[sample_target], s=50);
```

## Generate Linkage Matrix

```python
z = linkage(sample_data) 
```

```python
dendrogram(z,  p=12, leaf_rotation=45., leaf_font_size=10.)
plt.title('Truncated Hierarchical Clustering Dendrograms')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')

plt.show()
```

## Agglomerative Clustering

```python
clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')

clustering.fit(sample_data)

clustering.labels_
```

```python
plt.scatter(sample_data[:, 0], sample_data[:, 1],  c=color_theme[clustering.labels_], s=50);
```
